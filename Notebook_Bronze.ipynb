# **2.CRM_SOURCE _FILE**

```python
def table_create(csv_files):
    for file in csv_files:
        df = (
            spark.read.option("header", "true")
            .option("inferSchema", "true")
            .csv("/Volumes/workspace/bronze/source_systems/source_crm/" + file)
        )
        df.write.mode("overwrite").saveAsTable("workspace.bronze." + 'crm_' + file[:-4])
table_create(csv_files)
```

# **2.ERP_SOURCE _FILE**

```python
def get_letters_till_underscore(s):
    return ''.join([c for c in s if c != '_' and s.index(c) < s.index('_')])
```

```python
def table_create(source):
    for file in source:
        if file == "PX_CAT_G1V2.csv":
            df = (
            spark.read.option("header", "true")
            .option("inferSchema", "true")
            .csv("/Volumes/workspace/bronze/source_systems/source_erp/" + file)
        )
            df.write.mode("overwrite").saveAsTable("workspace.bronze." + 'erp_' + 'category')

        else:
            df = (
                spark.read.option("header", "true")
                .option("inferSchema", "true")
                .csv("/Volumes/workspace/bronze/source_systems/source_erp/" + file)
            )
            df.write.mode("overwrite").saveAsTable("workspace.bronze." + 'erp_' + get_letters_till_underscore(file))
table_create(csv_source_erp)
```
